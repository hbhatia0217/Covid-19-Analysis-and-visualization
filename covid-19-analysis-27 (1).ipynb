```python

plt.figure(figsize=(14,10))
base_color = sb.color_palette()[0]
sb.countplot(data = df, x = 'continent', color = base_color)

# add annotations
n_points = df.shape[0]
cat_counts = df['continent'].value_counts()
locs, labels = plt.xticks() # get the current tick locations and labels

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):

    # get the text property for the label to get the correct count
    count = cat_counts[label.get_text()]
    pct_string = '{:0.1f}%'.format(100*count/n_points)

    # print the annotation just below the top of the bar
    plt.text(loc, count-8, pct_string, ha = 'center', color = 'w')
```

# Explanatory Analysis with multivariated analysis

1st Analysis: Perform polishing and cleaning steps for the following multivariate plot, 
where the relationship between total_cases and new_cases are depicted for 
the India and Mexico types. 

In below analysis, it shows total cases in Mexico were increasing at much faster pace in April than in India.  
However, by August of this year total cases in India were almost 2 million 
whereas total cases in Mexico were close to 500000 which is 150% more 
than cases in India

```python
# data processing: isolate only fairy and dragon type Pokemon
df =  pd.read_csv('owid-covid-data.csv')
df['month'] = pd.DatetimeIndex(df['date']).month
place_cols = ['continent','location']
non_place_cols = df.columns.difference(place_cols)
place_types = df.melt(id_vars = non_place_cols, value_vars = place_cols, 
                          var_name = 'place_type', value_name = 'place').dropna()
#print(pkmn_types)

place_sub = place_types.loc[place_types['place'].isin(['India','Mexico'])]

place_colors = {'India': '#ee99ac', 'Mexico': '#7038f8'}
```

```python
# MODIFY THIS CODE
plt.figure(figsize = [16, 8])
g = sb.FacetGrid(data = place_sub, hue = 'place', height = 5,palette = place_colors)
g.map(plt.scatter,'total_cases','month')
g.set(xscale = 'log')
x_ticks = [1000, 3000, 10000, 30000, 150000, 500000]
g.set(xticks = x_ticks, xticklabels = x_ticks)

# add labels and titles
g.set_xlabels('total_cases')
g.set_ylabels('month')
plt.title('Comparison Of Total Cases From March ill August between India and Mexico')
plt.legend(['India', 'Mexico'], title = 'Covid Cases by Months between India and Mexico')
```

```python
# Created heatmap to show which continent has which continent 
#has most deaths between Jan till August. Based on below analysis, month of August had most 
#total number of deaths in the world with the mean of 366K followed by June.

sb.set(font_scale=.75)
df["total_deaths"] = pd.to_numeric(df["total_deaths"])
df["new_cases"] = pd.to_numeric(df["new_cases"])
#cases = df.loc[df['continent'].isin(['North America','Asia'])]
case_means = df.groupby(['continent', 'month']).mean()['total_deaths']
case_means = case_means.reset_index(name = 'total_deaths_avg')
case_means = case_means.pivot(index = 'month', columns = 'continent',
                            values = 'total_deaths_avg')
sb.heatmap(case_means, annot = True, fmt = '.3f',
           cbar_kws = {'label' : 'mean(total_deaths)'})
```
